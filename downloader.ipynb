```python
from diffusers import StableDiffusionPipeline, UNet2DConditionModel, AutoencoderKL
from transformers import CLIPTextModel, CLIPTokenizer
import torch
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
pretrained_model_name_or_path = "stabilityai/stable-diffusion-2"
save_directory = "/content/drive/MyDrive/Drexel/CS614 Applications of Machine Learning/CS614 Final Project/model"
```

```python
# Load the pipeline with caching to the save directory
pipeline = StableDiffusionPipeline.from_pretrained(
    pretrained_model_name_or_path,
    revision="fp16",
    torch_dtype=torch.float16,
    cache_dir=save_directory
).to("cuda")
```

```python
# Similarly, load individual components with the cache_dir set
text_encoder = CLIPTextModel.from_pretrained(
    pretrained_model_name_or_path,
    subfolder="text_encoder",
    cache_dir=save_directory
)
vae = AutoencoderKL.from_pretrained(
    pretrained_model_name_or_path,
    subfolder="vae",
    cache_dir=save_directory
)
unet = UNet2DConditionModel.from_pretrained(
    pretrained_model_name_or_path,
    subfolder="unet",
    cache_dir=save_directory
)
tokenizer = CLIPTokenizer.from_pretrained(
    pretrained_model_name_or_path,
    subfolder="tokenizer",
    cache_dir=save_directory
)
```

```python
# Save the full pipeline (if supported)
pipeline.save_pretrained(save_directory)

# Alternatively, save each component separately
text_encoder.save_pretrained(f"{save_directory}/text_encoder")
vae.save_pretrained(f"{save_directory}/vae")
unet.save_pretrained(f"{save_directory}/unet")
tokenizer.save_pretrained(f"{save_directory}/tokenizer")
```
